```python
import pandas as pd
```

```python
sales_report ={
     'products_id'  :[23,96,97,15,87],
           'name'  :['computer','python workout','pandas workout','banana','sandwich'],
    'wholesale_price':[500,35,35,0.5,3],
    'retail_price':[1000,75,75,1,5],
    'sales': [100,1000,500,200,300]
}
#creating dataframe
df =pd.DataFrame(sales_report)
```

```python
df
```

```python
#setting name as index
df.set_index('name',inplace=True)
```

```python
df
```

```python
#adding the net_revenue_per product before the sales tax introduced
df['current_income']=(df['retail_price']-df['wholesale_price'])* df['sales']
df
```

```python
#calculating current income after a 15% sales tax is introduced.
df['current_income_15%']=(df['retail_price']-df['wholesale_price'])*(df['sales']-15/100*df['sales'])
df
```

```python
#calculating current income after a 20% sales tax is introduced
df['current_income_20%']=(df['retail_price']-df['wholesale_price'])*(df['sales']-20/100*df['sales'])
df
```

```python
#calculating current income after a 25% sales tax is introduced
df['current_income_25%']=(df['retail_price']-df['wholesale_price'])*(df['sales']-25/100*df['sales'])
df
```
